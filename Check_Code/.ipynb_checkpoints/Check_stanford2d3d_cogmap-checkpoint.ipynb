```python
import os
import re
import cv2
import json
import array
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import OpenEXR
import Imath
from mpl_toolkits.mplot3d import Axes3D
from scipy import ndimage
from collections import defaultdict
import matplotlib.patches as mpatches
```

```python
# ç”Ÿæˆè®¤çŸ¥å›¾
def generate_cognitive_map(instance_mask, xyz_map, id_to_class, grid_size=(20, 20)):
    """
    å‚æ•°:
        instance_mask: H x W çš„ numpy æ•°ç»„ï¼Œå®ä¾‹åˆ†å‰²ç°åº¦å›¾ï¼Œæ¯ä¸ªåƒç´ å€¼æ˜¯å®ä¾‹ID
        xyz_map: H x W x 3 çš„ numpy æ•°ç»„ï¼Œå­˜å‚¨æ¯ä¸ªåƒç´ çš„ 3D åæ ‡
        id_to_class: å­—å…¸ï¼Œ{instance_id: class_name}
        grid_size: ç©ºé—´è®¤çŸ¥å›¾åˆ’åˆ†çš„ç½‘æ ¼å°ºå¯¸ (rows, cols)
    è¿”å›:
        ç©ºé—´è®¤çŸ¥å›¾çš„å¯è§†åŒ–
    """
    # æå–æ¯ä¸ªå®ä¾‹çš„ 3D ä¸­å¿ƒ
    instance_positions = {}
    for instance_id in np.unique(instance_mask):
        if instance_id == 0:  # èƒŒæ™¯å¿½ç•¥
            continue
        mask = instance_mask == instance_id
        coords = xyz_map[mask]
        coords = coords[~np.isnan(coords).any(axis=1)]  # å»é™¤æ— æ•ˆå€¼
        if len(coords) == 0:
            continue
        center = np.median(coords, axis=0)  # ä½¿ç”¨ä¸­ä½æ•°é¿å…ç¦»ç¾¤å€¼
        instance_positions[instance_id] = center

    # è®¡ç®—å…¨ä½“çš„ç©ºé—´èŒƒå›´ç”¨äºç½‘æ ¼å½’ä¸€åŒ–
    all_centers = np.array(list(instance_positions.values()))
    min_xyz = np.min(all_centers, axis=0)
    max_xyz = np.max(all_centers, axis=0)

    # æ„å»ºç©ºç™½ç½‘æ ¼å›¾
    grid = [[[] for _ in range(grid_size[1])] for _ in range(grid_size[0])]
    
    for inst_id, center in instance_positions.items():
        # æŠŠç©ºé—´å°ºåº¦å½’ä¸€åŒ–åˆ°ç½‘æ ¼ä¸Š
        normalized = (center - min_xyz) / (max_xyz - min_xyz + 1e-6)  # é¿å…é™¤ä»¥0
        grid_y = min(int(normalized[1] * grid_size[0]), grid_size[0] - 1)
        grid_x = min(int(normalized[0] * grid_size[1]), grid_size[1] - 1)
        class_name = id_to_class.get(inst_id, f"id{inst_id}")
        grid[grid_y][grid_x].append(class_name)

    # å¯è§†åŒ–ç©ºé—´è®¤çŸ¥å›¾
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.set_xlim(0, grid_size[1])
    ax.set_ylim(0, grid_size[0])
    ax.invert_yaxis()
    ax.set_xticks(np.arange(0, grid_size[1]+1))
    ax.set_yticks(np.arange(0, grid_size[0]+1))
    ax.grid(True)

    # æ·»åŠ æ¯ä¸ªç‰©ä½“çš„ä½ç½®
    for i in range(grid_size[0]):
        for j in range(grid_size[1]):
            items = grid[i][j]
            if items:
                label = '\n'.join(sorted(set(items)))
                ax.text(j + 0.5, i + 0.5, label, ha='center', va='center', fontsize=8, wrap=True)

    ax.set_title("Spatial Cognitive Map", fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
# è¯»å–exræ–‡ä»¶
def read_exr( image_fpath ):
    """ Reads an openEXR file into an RGB matrix with floats """
    f = OpenEXR.InputFile( image_fpath )
    dw = f.header()['dataWindow']
    w, h = (dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1)    
    im = np.empty( (h, w, 3) )

    # Read in the EXR
    FLOAT = Imath.PixelType(Imath.PixelType.FLOAT)
    channels = f.channels( ["R", "G", "B"], FLOAT )
    for i, channel in enumerate( channels ):
        im[:,:,i] = np.reshape( array.array( 'f', channel ), (h, w) )
    return im

""" Label functions """
def load_labels( label_file ):
    """ Convenience function for loading JSON labels """
    with open( label_file ) as f:
        return json.load( f )

def parse_label( label ):
    """ Parses a label into a dict """
    res = {}
    clazz, instance_num, room_type, room_num, area_num = label.split( "_" )
    res[ 'instance_class' ] = clazz
    res[ 'instance_num' ] = int( instance_num )
    res[ 'room_type' ] = room_type
    res[ 'room_num' ] = int( room_num )
    res[ 'area_num' ] = int( area_num )
    return res

def get_index(color):
    """ Parse a color as a base-256 number and return the index """
    r, g, b = map(int, color)  # ğŸ‘ˆ è½¬æ¢ä¸º Python int ç±»å‹ï¼Œé¿å… uint8 æº¢å‡º
    return r * 256 * 256 + g * 256 + b
```

```python
# ç¡®å®šç›¸æœºä½ç½®
import numpy as np

def estimate_camera_position_from_equirectangular(xyz_image, depth_map, rgb_image):
    """
    æ ¹æ® equirectangular å…¨æ™¯å›¾åƒã€æ·±åº¦å›¾å’Œ 3D ç‚¹å›¾ä¼°è®¡ç›¸æœºåœ¨ä¸–ç•Œåæ ‡ç³»ä¸­çš„ä½ç½®ã€‚

    å‚æ•°:
        xyz_image: (H, W, 3) numpy æ•°ç»„ï¼Œæ¯ä¸ªåƒç´ çš„ä¸–ç•Œåæ ‡ [X, Y, Z]
        depth_map: (H, W) numpy æ•°ç»„ï¼Œæ¯ä¸ªåƒç´ çš„æ·±åº¦ï¼ˆä»ç›¸æœºå‡ºå‘åˆ°è¡¨é¢çš„è·ç¦»ï¼‰
        rgb_image: (H, W, 3) numpy æ•°ç»„ï¼Œç”¨äºè·å–å›¾åƒå°ºå¯¸

    è¿”å›:
        camera_position: (3,) numpy æ•°ç»„ï¼Œç›¸æœºåœ¨ä¸–ç•Œåæ ‡ç³»ä¸­çš„ä¼°è®¡ä½ç½®
    """
    H, W = rgb_image.shape[:2]

    # Step 1: ç”Ÿæˆæ¯ä¸ªåƒç´ çš„çƒé¢æ–¹å‘ï¼ˆå•ä½å‘é‡ï¼‰
    # è®¡ç®—æ¯ä¸ªåƒç´ çš„çƒé¢è§’åº¦ï¼šÎ¸ï¼ˆæ°´å¹³ï¼‰ã€Ï†ï¼ˆå‚ç›´ï¼‰
    u = np.linspace(0, 1, W, endpoint=False)
    v = np.linspace(0, 1, H, endpoint=False)
    uu, vv = np.meshgrid(u, v)

    theta = 2 * np.pi * (uu - 0.5)  # æ°´å¹³ [-Ï€, Ï€]
    phi = np.pi * (vv - 0.5)       # å‚ç›´ [-Ï€/2, Ï€/2]

    # å¯¹åº”å•ä½æ–¹å‘å‘é‡ï¼šä»ç›¸æœºå‡ºå‘
    ray_dirs = np.stack([
        np.cos(phi) * np.sin(theta),  # x
        np.sin(phi),                  # y
        np.cos(phi) * np.cos(theta)   # z
    ], axis=-1)  # shape: (H, W, 3)

    # Step 2: æå–æœ‰æ•ˆåƒç´ ç‚¹ï¼ˆé NaN ä¸” depth åˆç†ï¼‰
    valid_mask = (~np.isnan(xyz_image).any(axis=-1)) & (~np.isnan(depth_map)) & (depth_map > 0)
    valid_xyz = xyz_image[valid_mask]
    valid_depth = depth_map[valid_mask]
    valid_dirs = ray_dirs[valid_mask]

    if len(valid_xyz) == 0:
        raise ValueError("æ²¡æœ‰æœ‰æ•ˆåƒç´ ç”¨äºä¼°è®¡ç›¸æœºä½ç½®ï¼Œè¯·æ£€æŸ¥è¾“å…¥ã€‚")

    # Step 3: åˆ©ç”¨å…¬å¼åæ¨å‡ºç›¸æœºä½ç½®ï¼š
    # cam = p_i - depth_i * v_dir_i
    cam_candidates = valid_xyz - valid_depth[:, np.newaxis] * valid_dirs

    # Step 4: å¤šç‚¹å–å¹³å‡ï¼Œå‡å°‘è¯¯å·®å½±å“ï¼ˆä¾‹å¦‚æ·±åº¦é‡åŒ–è¯¯å·®ã€é®æŒ¡ï¼‰
    camera_position = np.mean(cam_candidates, axis=0)

    return camera_position
```
